## Mengatasi Missing Values dengan Drop Data

### 1. Hapus Data yang kosong (NaN)

```python
df = pd.read_csv('./Data.csv')
df.dropna(inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Baru

### 2. Ganti Data yang kosong (NaN) menjadi :

### <ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
df = pd.read_csv('./Data.csv')
df['Umur'].fillna(50, inplace=True)
df['Gaji'].fillna(25000000, inplace=True)
df
```

## Mengatasi Missing Values dengan Nilai Mean

## 3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mean();
mean_gaji = df['Gaji'].mean();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)
df
```

## Mengatasi Missing Values dengan Nilai Median

## 4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].median();
mean_gaji = df['Gaji'].median();

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Missing Values dengan Nilai Mode

## 5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df = pd.read_csv('./Data.csv')
mean_umur = df['Umur'].mode()[0];
mean_gaji = df['Gaji'].mode()[0];

df['Umur'].fillna(mean_umur, inplace=True)
df['Gaji'].fillna(mean_gaji, inplace=True)

df
```

## Mengatasi Wrong Data dengan Ubah Data

## 6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
df = pd.read_csv('./Data.csv')

for x in df.index:
    if df.loc[x, 'Umur'] > 120:
        df.loc[x, 'Umur'] = 33

df
```

# Tugas Praktikum

```python
import pandas as pd
import datetime as dt

df_customers = pd.read_csv('./customers.csv')
df_customers.head()
df_products = pd.read_csv('./products.csv')
df_products.head()
df_gasstations = pd.read_csv('./gasstations.csv')
df_gasstations.head()
df_yearmonth = pd.read_csv('./yearmonth.csv')
df_yearmonth.head()
df_transactions = pd.read_csv('./transactions_1k.csv')
df_transactions.head()
```

```python
df_merged = pd.merge(df_transactions, df_customers,
                     on='CustomerID', how='left')

df_merged = pd.merge(df_merged, df_gasstations, on='GasStationID', how='left')
df_merged = pd.merge(df_merged, df_products, on='ProductID', how='left')
df_merged['Date'] = pd.to_datetime(df_merged['Date'])
df_merged['YearMonth'] = df_merged['Date'].dt.strftime('%Y%m')
df_yearmonth['Date'] = df_yearmonth['Date'].astype(str)
df_merged = pd.merge(
    df_merged,
    df_yearmonth,
    left_on=['CustomerID', 'YearMonth'],
    right_on=['CustomerID', 'Date'],
    how='left'
)

df_merged.drop(columns=['Date_y', 'YearMonth'], inplace=True)
df_merged.rename(columns={'Date_x': 'Date'}, inplace=True)
df_merged['Date'] = df_merged['Date'].astype(str)

df_merged.head()
```

## 1. Table list Top 5 Customers yang memiliki nilai transaksi paling banyak

```python
merged = pd.merge(df_transactions, df_customers, on="CustomerID")
summary = merged.groupby("CustomerID").agg(
    Amount=('TransactionID', 'count'),
    Sales=('Price', 'sum')
).sort_values(by="Sales", ascending=False).head(5)
summary.index.name = "Customer ID"
summary
```

## 2. list Top 5 Gas stations yang memiliki nilai transaksi paling banyak



```python
merged = pd.merge(df_transactions, df_customers, on="CustomerID")
summary = merged.groupby("CustomerID").agg(
    Amount=('TransactionID', 'count'),
    Sales=('Price', 'sum')
).sort_values(by="Sales", ascending=False).head(5)

summary.index.name = "Customer ID"
summary
```

## 3. list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak

```python
merged = pd.merge(df_transactions, df_products, on="ProductID")
summary = merged.groupby("Description").agg(
    Amount=("Amount", "sum"),
    Sales=("Price", "sum"),
).sort_values(by="Sales", ascending=False).head(5)
summary.index.name = "Products"
summary
```

## 4. deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset transaction_1k

```python
df_transactions["Date"] = pd.to_datetime(df_transactions["Date"])
filtered_df = df_transactions[df_transactions["Date"].dt.day.isin([23, 24, 25, 26])]
stat_funcs = ['count', 'mean', 'std', 'min',
              lambda x: x.quantile(0.25),
              'median',
              lambda x: x.quantile(0.75),
              'max']
summary = filtered_df.groupby("Date").agg({
    'Amount': stat_funcs,
    'Price': stat_funcs
})

stat_names = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
summary.columns = pd.MultiIndex.from_tuples(
    [(col[0], stat_names[i % 8]) for i, col in enumerate(summary.columns)]
)

summary.rename(columns={"Price": "Sales"}, inplace=True)
summary.columns.names = ['Metric', 'Statistic']
summary
```

## 5. waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi

```python
df_transactions['Amount'] = pd.to_numeric(
    df_transactions['Amount'], errors='coerce')
df_transactions['Price'] = pd.to_numeric(
    df_transactions['Price'], errors='coerce')
df_transactions['Datetime'] = pd.to_datetime(
    df_transactions['Date'].astype(
        str) + ' ' + df_transactions['Time'].astype(str),
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)
df_transactions['Date'] = df_transactions['Datetime'].dt.strftime('%Y-%m-%d')
df_transactions['Hour'] = df_transactions['Datetime'].dt.hour

result = df_transactions.groupby(['Date', 'Hour']).agg({
    'Amount': 'sum',
    'Price': 'sum'
}).reset_index()
result
```
