```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Pandas

### a. Read

```python
df = pd.read_csv('/content/airbnb.csv')
print("Head DataFrame:\n", df.head())
```

### b. Series

```python
print("\nSeries - Nama Host:\n", df['host_name'].head())
```

### c. DataFrame

```python
print("\nInfo DataFrame:")
df_info = df.info()
print(df_info)
print("\nDeskripsi Statistik:")
print(df.describe())
```

### d. Manipulation

```python
df_filtered = df[df['price'] < 100]
print("\nData dengan harga < 100:\n", df_filtered[['name', 'price']].head())
```

### e. Plotting

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, color='cyan', edgecolor='black')
plt.title('Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Jumlah Listing')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 2. Matplotib

### a. Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room_type', y='price')
plt.ylim(0, 500)
plt.title('Distribusi Harga Berdasarkan Tipe Kamar')
plt.xlabel('Tipe Kamar')
plt.ylabel('Harga ($)')
plt.tight_layout()
plt.show()
```

### b. Customization

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
colors = sns.color_palette("Set2")

order = df['neighbourhood_group'].value_counts().index
sns.countplot(data=df, x='neighbourhood_group', order=order, palette=colors)

plt.title('Distribusi Jumlah Listing Airbnb per Wilayah NYC', fontsize=16, fontweight='bold')
plt.xlabel('Wilayah (Neighbourhood Group)', fontsize=12)
plt.ylabel('Jumlah Listing', fontsize=12)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

### c. Subplot

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')

df = df.dropna(subset=['price'])

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['price'], bins=50, kde=True, color='skyblue')
plt.title('Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Jumlah Listing')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['price'], color='salmon')
plt.title('Boxplot Harga')
plt.xlabel('Harga')

plt.tight_layout()
plt.show()
```

### d. Diagram

```python
print("Kolom tersedia:", df.columns)
kolom_harga = 'price'

df[kolom_harga] = pd.to_numeric(df[kolom_harga], errors='coerce')

df = df.dropna(subset=[kolom_harga])

plt.figure(figsize=(10, 6))
sns.histplot(df[kolom_harga], bins=50, kde=True)
plt.title('Distribusi Harga Airbnb')
plt.xlabel('Harga')
plt.ylabel('Jumlah Listing')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(x=df[kolom_harga])
plt.title('Boxplot Harga Airbnb')
plt.xlabel('Harga')
plt.grid(True)
plt.show()
```

### 3. Data Analysis & Manipulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("airbnb.csv")

print("5 Data Teratas:")
print(df.head())

print("\nInfo Dataset:")
print(df.info())

print("\nStatistik Deskriptif:")
print(df.describe(include='all'))

print("\nJumlah Nilai Kosong per Kolom:")
print(df.isnull().sum())

columns_to_drop = ['id', 'name'] if 'id' in df.columns and 'name' in df.columns else []
df_clean = df.drop(columns=columns_to_drop)

df_clean.fillna({
    'reviews_per_month': 0,
    'last_review': 'Unknown'
}, inplace=True)

if 'room_type' in df.columns and 'price' in df.columns:
    avg_price = df.groupby('room_type')['price'].mean().sort_values(ascending=False)
    print("\nRata-rata Harga per Tipe Kamar:")
    print(avg_price)

    plt.figure(figsize=(8,5))
    sns.barplot(x=avg_price.index, y=avg_price.values, palette='viridis')
    plt.title("Rata-rata Harga Airbnb per Tipe Kamar")
    plt.ylabel("Harga Rata-rata")
    plt.xlabel("Tipe Kamar")
    plt.tight_layout()
    plt.show()

df_clean.to_csv("airbnb_clean.csv", index=False)
print("\nDataset bersih disimpan sebagai 'airbnb_clean.csv'")
```
